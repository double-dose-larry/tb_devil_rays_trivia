```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import tabulate
from functools import partial
%matplotlib inline
pd.set_option('display.max_rows', 300)
pd.set_option('display.max_columns', 30)
```

```python
def get_season_data(year):
    df = pd.read_html(f"https://widgets.sports-reference.com/wg.fcgi?css=1&site=br&url=%2Fteams%2FTBD%2F{year}-batting.shtml&div=div_team_batting")[0].query("Pos != 'Pos'")
    df["year"] = year
    return df

to_num = partial(pd.to_numeric, errors="ignore")

def ba(row):
    return row.H/row.AB

def obp(row):
    return (row.H + row.BB + row.HBP)/(row.AB + row.BB + row.HBP + row.SF)

def slg(row):
    singles = row.H - (row["2B"] + row["3B"] + row["HR"])
    return (singles + row["2B"] * 2 + row["3B"] * 3 + row["HR"] * 4)/row.AB
```

```python
devil_ray_years = range(1998,2008)
```

```python
df = pd.concat([ get_season_data(year) for year in devil_ray_years])
```

```python
# drop the year totals from the data
df = df.dropna(subset=["Rk"])
# make numeric
df = df.apply(to_num)
```

```python
# columns to sum up
cols = ['G', 'PA', 'AB', 'R', 'H', '2B', '3B', 'HR', 'RBI', 'SB', 'CS', 'BB', 'SO', 'TB', 'GDP', 'HBP', 'SH', 'SF', 'IBB']
```

```python
totals = df.groupby(["Name"])[cols].sum().sort_values("3B", ascending = False)
```

```python
totals["ba"] = totals.apply(ba, axis=1)
totals["obp"] = totals.apply(obp, axis=1)
totals["slg"] = totals.apply(slg, axis=1)
totals["ops"] = totals.slg + totals.obp
totals["ab/so"] = totals.AB/totals.SO
```
